```python
import tensorflow as tf
tf.__version__
```

### Create a vector, scalar, matrix and tensor with values of your choosing using tf.constant().

```python
scalar = tf.constant(10)
scalar, scalar.ndim
```

```python
vector = tf.constant([10, 20, 30])
vector, vector.ndim
```

```python
matrix = tf.constant([[10, 20], [20, 30], [30, 40]])
matrix, matrix.ndim
```

### Find the shape, rank and size of the tensors you created in 1.


```python
tf.shape(scalar), tf.rank(scalar), tf.size(scalar)
```

```python
tf.shape(vector), tf.rank(vector), tf.size(vector)
```

```python
tf.shape(matrix), tf.rank(matrix), tf.size(matrix)
```

### Create two tensors containing random values between 0 and 1 with shape [5, 300].

```python
random_tensor1 = (tf.random.uniform(shape = [5, 300], minval=0, maxval=1))
random_tensor2 = (tf.random.uniform(shape = [5, 300], minval=0, maxval=1))
```

```python
random_tensor1, random_tensor2
```

### Multiply the two tensors you created in 3 using matrix multiplication.


```python
tf.multiply(random_tensor1, random_tensor2)
```

### Multiply the two tensors you created in 3 using dot product.


```python
tf.matmul(random_tensor1, tf.transpose(random_tensor2))
```

### Create a tensor with random values between 0 and 1 with shape [224, 224, 3].


```python
random_val = tf.random.uniform(shape=[224, 224, 3], minval=0, maxval=1)
random_val
```

```python
tf.reduce_min(random_val, axis=0)
```

```python
tf.reduce_max(random_val, axis = 0)
```

### Created a tensor with random values of shape [1, 224, 224, 3] then squeeze it to change the shape to [224, 224, 3].

```python
random_tensor = tf.random.uniform(shape=[1, 224, 224, 3])
random_tensor, random_tensor.shape
```

```python
tf.squeeze(random_tensor)
```

### Create a tensor with shape [10] using your own choice of values, then find the index which has the maximum value.


```python
ten_shape = tf.constant([1, 2, 3, 4, 5, 20, 7, 8, 9, 10])
ten_shape
```

```python
tf.argmax(ten_shape)
```

### One-hot encode the tensor you created in 9.

```python
tf.one_hot(ten_shape, depth = 10)
```
