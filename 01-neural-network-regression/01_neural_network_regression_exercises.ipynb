```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = tf.range(0, 500, 10)
y = X + 10
```

```python
X, y
```

```python
plt.scatter(X, y)
```

```python
X_train = np.array(X[:40]).reshape(-1, 1)
y_train = np.array(y[:40]).reshape(-1, 1)

X_test = np.array(X[40:]).reshape(-1, 1)
y_test = np.array(y[40:]).reshape(-1, 1)
```

```python
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
plt.figure(figsize=[8, 5])
plt.scatter(X_train, y_train, c="g", label="Train Data")
plt.scatter(X_test, y_test, c = "b", label="Test Data")
plt.legend();
```

```python
X_train, y_train
```

```python
tf.random.set_seed(42)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(50),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1),
])

model.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001),
    metrics = ["mae"]
)

model.fit(X_train, y_train, epochs = 200)
```

```python
X_test
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
def plot_prediction(
    train_data=X_train,
    train_labels=y_train,
    test_data=X_test,
    test_labels=y_test,
    predction=y_pred
    ):

  plt.figure(figsize=(6, 4))
  plt.scatter(train_data, train_labels, c="b", label="Training Data")
  plt.scatter(test_data, test_labels, c="g", label="Testing Data")
  plt.scatter(test_data, predction, c="r", label="Predictions")
  plt.legend();
```

```python
plot_prediction()
```
